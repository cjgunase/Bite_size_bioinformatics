# Load Genotype Data

#### Load and match the sample IDs in genotype data with sample ID in methylation data

```r
list.files("./VCF/Genotype_post_imputation/")
sample_id_table1 <- read.table("./misc_files/sample_id_table1.txt",header=T)
sample_id_table2 <- read.table("./misc_files/subjectIDs.txt",header=T)
SAMPLE_IDs<- merge(sample_id_table1,sample_id_table2,by.x = "ID2",by.y = "blindid")
head(SAMPLE_IDs)
```

# Sanity Check
## Pick a CoRSIV region and plot DNA methylation vs Genotype

```r
CoRSIV_ID = "e2a298d0"
methylation_data[methylation_data$CoRSIV_ID==CoRSIV_ID,]
```

```r
CHR="chr8"
vcf_chr <- read.vcfR(file = paste0("./VCF/Genotype_post_imputation/",CHR,".subset.dose.vcf.gz"))
gt_chr <- extract.gt(vcf_chr,element = 'GT')
SAMPLE_IDs <- SAMPLE_IDs[match(colnames(gt_chr), SAMPLE_IDs$ID),]#This will endup with sample IDs that will be used in this analysis
colnames(gt_chr) <- SAMPLE_IDs$ID1
gt_chr <- data.frame(gt_chr)
gt_chr$snp_id <- rownames(gt_chr)
gt_chr <- separate(data=gt_chr,col=snp_id,into = c("chr","loc","REF","ALT"),sep = ":" )
```

# Extract genotype in the haplotype region

```r
snps_unformatted <- strsplit(as.character(CoRSIV_Haplotype_BED[CoRSIV_Haplotype_BED$V10==CoRSIV_ID,]$V6),split = "\\|")[[1]]
snps_formated <- str_replace_all(str_replace(snps_unformatted, "_b38", ""),"_",":")
snps_formated <- intersect(rownames(gt_chr),snps_formated)
snps_gt_data <- gt_chr[rownames(gt_chr) %in% snps_formated,]
snps_gt_data <- snps_gt_data[,1:length(SAMPLE_IDs$ID1)]
snps_gt_data
```

```r
gt_vector <- as.factor(sapply(snps_gt_data, function(x) paste(as.character(x),collapse = '--')))
methy_vector <- as.numeric(methylation_data[methylation_data$CoRSIV_ID==CoRSIV_ID,][colnames(snps_gt_data)[1:length(SAMPLE_IDs$ID1)]])
gt_methy <- data.frame(gt_vector,methy_vector)
ggboxplot(gt_methy, x = "gt_vector", y = "methy_vector",
           add = "dotplot")+rotate_x_text(angle = 25)+ylab("% methylation")+xlab("genotype")
```
